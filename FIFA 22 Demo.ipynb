```r
#Import data using read.csv function
data <- read.csv("Fifa 22 Data.csv", stringsAsFactors=FALSE, encoding="UTF-8")
head(data)
```

```r
#Question 1
q1 <- data[, c(1, 2, 4, 8, 9)]
head(q1, n=8)
```

```r
#Question 2
mean_age <- mean(q1$Age)
mean_overall <- mean(q1$Overall)
mean_age; mean_overall
```

```r
#Question 3
q3 <- q1[q1$Nationality=="France", ]
```

```r
#Method 1 - Use mean function to calculate average overall and use the order function to identify the highest rated player(s)
mean_overall_france <- mean(q3$Overall)
print(mean_overall_france)
```

```r
top_players_france <- q3[order(-q3$Overall), ]
print(top_players_france)
```

```r
#Method 2 - Use the summary function to generate all required values
summary(q3)
```

```r
#Import data
fifa_df <- read.csv("data.csv")
head(fifa_df)
```

```r
#Question 1

#Which player has the highest goalkeeper rating (GKRating)? Ensure your code shows both the player name and his GKRating.

#Step 1 - Create two data frame with colums of interest (columns 3 and 90)
q1_temp <- fifa_df[ , c(3, 90)] #OR q1_temp <- fifa_df[ , c("FullName", "GKRating")]

#Step 2 - Sort data in descending order by GKRating
q1_sorted <- q1_temp[order(-q1_temp$GKRating), ] #The negative sign indicates we're sorting in descending order

#Step 3 - Retrieve first row of q1_sorted
head(q1_sorted, n=1)
```

```r
#Question 2

#Create a data frame that only includes players from Canada. Name it can_df. How many total players are in this data frame?

#Step 1 - Create a data frame that only includes Canadian players
q2 <- fifa_df[fifa_df$Nationality=="Canada", ] #Remember you have to use a double equals sign

#Step 2 - Compute number of rows in data frame
nrow(q2)
```

```r
#Question 3

#Create a data frame for players older than 30 and whose overall is 90 or higher. How many players are in this data frame?
#Note that older than 30 means 31 and older (hence, > 30)

#Step 1 - Create a data frame based on the requested conditions
q3 <- fifa_df[(fifa_df$Age>30 & fifa_df$Overall>=90), ]

#Step 2 - Compute number of rows in data frame
nrow(q3)
```

```r
#Appendix

#A few students have asked if there's a way to print all columns of a data frame. When I call head(fifa_df), notice how some
#columns become truncuated, as indicated by the ellipsis below:
head(fifa_df)
```

```r
#You change change the default print settings as follows
options(repr.matrix.max.cols=150, rep.matrix.max.rows=200) #Tells R to print up to 150 cols and up to 200 rows

#Now, when I call head(fifa_df), the ellipsis is gone
head(fifa_df)
```

```r
#Extracting elements from vectors

#Let's create a new variable called "age" for the "Age" column of our FIFA data frame
age <- data$Age

#We could print all the ages in the vector by typing age or using the print function, but let's say we were only interested
#in the 11th element of the vector

#We could extract it as follows
age[11]

#Similarly, if we wanted the first element, we could extract it as 
age[1]

#If we wanted the last element of the vector, we could extract it as
age[length(age)]

#Alternatively, we could input the number that corresponds to the last element of the vector as follows
age[19260]
```

```r
#Extracting elements from lists

#Create and print a list called test_list. The colon operator allows you to specify ranges of values (e.g. 7:11 results in
#7, 8, 9, 10, 11)
test_list <- list(7:11, 0:4)
test_list
```

```r
#Extract first component of the list (i.e. 7 through 11)
test_list[[1]]
```

```r
#Extract third element of first part of list

#Method 1
first_component <- test_list[[1]]
first_component[3]
```

```r
#Extract third element of first part of list

#Method 2
test_list[[1]][3] #First, we select the list (either 1 or 2) using double square brackets. Then, we select the third element
```

```r
#Extracting elements from matrices

#Create test matrix by combining two vectors
vector1 <- c(1:5)
vector2 <- c(8:12)

#Use cbind function to create test_matrix variable
test_matrix <- cbind(vector1, vector2)

#Confirm we have a matrix and print it
class(test_matrix)
print(test_matrix)
```

```r
#Let's extrac the 4 element of the second column (11)
test_matrix[4, 2] #The first argument is the row number, while the second argument is the column number
```

```r
#Extracting elements from arrays

#Create test array with two matrices, each 3 by 3
test_array <- array(c(10:19), dim=c(3, 3, 2))
print(test_array)
```

```r
#Let's extract the last element of the first column of the second array (value is 11)
test_array[3, 1, 2]
```
